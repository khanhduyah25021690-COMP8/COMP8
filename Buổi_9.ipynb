```python
# W9A1:
import sys

def merge_sorted_lists(a, b):
    i = j = 0
    res = []

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            res.append(a[i])
            i += 1
        else:
            res.append(b[j])
            j += 1

    # Append phần còn lại (nếu có)
    if i < len(a):
        res.extend(a[i:])
    if j < len(b):
        res.extend(b[j:])

    return res

def parse_line(line: str):
    line = line.strip()
    if not line:
        return []
    return list(map(int, line.split()))

def main():
    lines = sys.stdin.read().splitlines()

    line1 = lines[0] if len(lines) >= 1 else ""
    line2 = lines[1] if len(lines) >= 2 else ""

    a = parse_line(line1)
    b = parse_line(line2)

    merged = merge_sorted_lists(a, b)
    print(*merged)

if __name__ == "__main__":
    main()
```

```python
# W9A2
import heapq

def kth_largest_in_sorted_matrix(matrix, k):
    n = len(matrix)
    m = len(matrix[0])

    heap = []

    # Đưa phần tử đầu mỗi dòng vào heap
    for i in range(n):
        heapq.heappush(heap, (matrix[i][0], i, 0))

    count = 0

    while heap:
        val, row, col = heapq.heappop(heap)
        count += 1

        if count == k:
            return val

        # Đẩy phần tử tiếp theo trong cùng dòng
        if col + 1 < m:
            heapq.heappush(heap, (matrix[row][col + 1], row, col + 1))

def main():
    n, m = map(int, input().split())
    matrix = [list(map(int, input().split())) for _ in range(n)]
    k = int(input())

    print(kth_largest_in_sorted_matrix(matrix, k))

if __name__ == "__main__":
    main()
```

```python
# W9A3:
import sys

def min_after_one_swap_exact(s: str) -> str:
    n = len(s)
    if n <= 1:
        return s  # không có gì để đổi

    # suffix_rightmost[i][d] = vị trí xuất hiện "bên phải xa nhất" của digit d trong s[i:]
    suffix_rightmost = [[-1] * 10 for _ in range(n + 1)]

    for i in range(n - 1, -1, -1):
        suffix_rightmost[i] = suffix_rightmost[i + 1].copy()
        suffix_rightmost[i][ord(s[i]) - 48] = i

    # 1) Cố gắng tìm một swap làm số nhỏ hơn (tối ưu theo i rồi theo digit nhỏ nhất)
    for i in range(n):
        cur = ord(s[i]) - 48
        right = suffix_rightmost[i + 1]  # chỉ xét phía sau i

        for d in range(cur):  # d < cur
            if i == 0 and d == 0:
                continue  # tránh tạo số bắt đầu bằng 0
            j = right[d]
            if j != -1:
                a = list(s)
                a[i], a[j] = a[j], a[i]
                return "".join(a)

    # 2) Không có swap nào làm nhỏ hơn
    # Nếu có chữ số lặp lại => có thể đổi 2 chữ số giống nhau để giữ nguyên (nhỏ nhất)
    freq = [0] * 10
    for ch in s:
        freq[ord(ch) - 48] += 1
    if any(c >= 2 for c in freq):
        return s

    # Nếu tất cả khác nhau => số đang "tối ưu" theo constraint, đổi 2 chữ số cuối là ít tăng nhất
    a = list(s)
    a[-1], a[-2] = a[-2], a[-1]
    return "".join(a)

def main():
    data = sys.stdin.read().split()
    out = []
    for token in data:
        out.append(min_after_one_swap_exact(token.strip()))
    sys.stdout.write("\n".join(out))

if __name__ == "__main__":
    main()
```

```python
# W9A4:
import sys

def can_transform(a: str, b: str) -> bool:
    m = len(b)
    dp = [False] * (m + 1)
    dp[0] = True

    for ch in a:
        new = [False] * (m + 1)

        if ch.islower():
            # Xóa ch
            for j in range(m + 1):
                if dp[j]:
                    new[j] = True

            # Đổi ch thành IN HOA
            up = ch.upper()
            for j in range(m):
                if dp[j] and up == b[j]:
                    new[j + 1] = True
        else:
            # Chữ IN HOA: bắt buộc dùng
            for j in range(m):
                if dp[j] and ch == b[j]:
                    new[j + 1] = True

        dp = new

    return dp[m]

def main():
    # Đọc và loại bỏ dòng trống
    lines = [line.strip() for line in sys.stdin if line.strip() != ""]

    if not lines:
        return  # không có input → thoát an toàn

    q = int(lines[0])
    idx = 1
    out = []

    for _ in range(q):
        a = lines[idx]
        b = lines[idx + 1]
        idx += 2
        out.append("YES" if can_transform(a, b) else "NO")

    print("\n".join(out))

if __name__ == "__main__":
    main()
```

```python
# W9A5:
import sys

def parse_list(line: str):
    line = line.strip()
    if not line:
        return []
    return list(map(int, line.split()))

def is_contiguous_sublist(a, b):
    # a = list1, b = list2
    k, n = len(a), len(b)
    if k == 0:
        return True  # dãy rỗng coi như luôn là dãy con
    if k > n:
        return False

    for i in range(n - k + 1):
        if b[i:i + k] == a:
            return True
    return False

def main():
    # Lấy các dòng không rỗng
    lines = [ln.strip() for ln in sys.stdin.read().splitlines() if ln.strip() != ""]
    out = []

    # Mỗi test = 2 dòng
    for i in range(0, len(lines), 2):
        if i + 1 >= len(lines):
            break
        list1 = parse_list(lines[i])
        list2 = parse_list(lines[i + 1])
        out.append("YES" if is_contiguous_sublist(list1, list2) else "NO")

    print("\n".join(out))

if __name__ == "__main__":
    main()
```

```python
# W9A6:
import sys

def next_permutation(a):
    n = len(a)

    # 1) tìm i: a[i] < a[i+1]
    i = n - 2
    while i >= 0 and a[i] >= a[i + 1]:
        i -= 1

    # 2) nếu không có i => hoán vị lớn nhất, trả về nhỏ nhất
    if i < 0:
        a.reverse()
        return a

    # 3) tìm j: từ phải sang trái, a[j] > a[i]
    j = n - 1
    while a[j] <= a[i]:
        j -= 1

    # 4) swap i và j
    a[i], a[j] = a[j], a[i]

    # 5) đảo đuôi i+1..n-1
    left, right = i + 1, n - 1
    while left < right:
        a[left], a[right] = a[right], a[left]
        left += 1
        right -= 1

    return a

def main():
    data = sys.stdin.read().split()
    if not data:
        return

    n = int(data[0])
    a = list(map(int, data[1:1 + n]))

    ans = next_permutation(a)
    print(*ans)

if __name__ == "__main__":
    main()
```

```python
# W9A7:
import sys

def canPlaceFlowers(flowerbed, k):
    if k <= 0:
        return True

    n = len(flowerbed)
    count = 0

    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i - 1] == 0)
            right_empty = (i == n - 1) or (flowerbed[i + 1] == 0)

            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
                if count >= k:
                    return True

    return False

def parse_list_line(line: str):
    line = line.strip()
    if not line:
        return []
    return list(map(int, line.split()))

def main():
    # Lấy các dòng không rỗng
    lines = [ln.strip() for ln in sys.stdin.read().splitlines() if ln.strip() != ""]
    out = []

    # Mỗi test: 2 dòng (flowerbed) + (k)
    for i in range(0, len(lines), 2):
        if i + 1 >= len(lines):
            break
        flowerbed = parse_list_line(lines[i])
        k = int(lines[i + 1])

        out.append(str(canPlaceFlowers(flowerbed, k)))

    print("\n".join(out))

if __name__ == "__main__":
    main()
```

```python
# W9AA8:
import sys

def is_path_crossing(moves: str) -> bool:
    x, y = 0, 0
    visited = {(0, 0)}

    for ch in moves.strip():
        if ch == 'R':
            x += 1
        elif ch == 'L':
            x -= 1
        elif ch == 'U':
            y += 1
        elif ch == 'D':
            y -= 1

        if (x, y) in visited:
            return True
        visited.add((x, y))

    return False

def main():
    lines = [ln.strip() for ln in sys.stdin.read().splitlines() if ln.strip() != ""]
    out = []
    for moves in lines:
        out.append(str(is_path_crossing(moves)))
    print("\n".join(out))

if __name__ == "__main__":
    main()
```

```python
# W9A9:
def Median(nums):
    # nums có đúng 5 số nguyên
    a, b, c, d, e = nums

    # Helper: đảm bảo x <= y
    def order2(x, y):
        if x > y:
            return y, x
        return x, y

    # Helper: median của 3 số (không sort), đúng với 3 phép so sánh
    def median3(x, y, z):
        if x > y:
            x, y = y, x
        if y > z:
            y, z = z, y
        if x > y:
            x, y = y, x
        return y

    # 1) Sắp xếp trong từng cặp (a,b) và (c,d)
    a, b = order2(a, b)   # a <= b
    c, d = order2(c, d)   # c <= d

    # 2) Đảm bảo a <= c bằng cách hoán đổi cả cặp nếu cần
    if a > c:
        a, c = c, a
        b, d = d, b
        # vẫn giữ a<=b và c<=d

    # 3) Đảm bảo b <= d (để a là nhỏ nhất trong 4 số và d là lớn nhất trong 4 số)
    if b > d:
        b, d = d, b

    # Lúc này: a là nhỏ nhất của {a,b,c,d} và d là lớn nhất của {a,b,c,d}
    # Trung vị của 5 số chính là median của {b, c, e}
    return median3(b, c, e)
```

```python
# W9A10:
def findRadius(houses, heaters):
    if not houses:
        return 0
    if not heaters:
        # Không có lò sưởi mà có nhà thì không thể phủ; tùy đề có thể không có case này.
        # Ở đây trả về 0 cho an toàn, nhưng thực tế nên coi là "vô nghiệm".
        return 0

    houses.sort()
    heaters.sort()

    j = 0
    ans = 0

    for h in houses:
        # đẩy j sao cho heaters[j] là heater đầu tiên >= h (hoặc j == len)
        while j < len(heaters) and heaters[j] < h:
            j += 1

        # khoảng cách tới heater bên phải (nếu có)
        right = abs(heaters[j] - h) if j < len(heaters) else float('inf')
        # khoảng cách tới heater bên trái (nếu có)
        left = abs(h - heaters[j - 1]) if j > 0 else float('inf')

        best = right if right < left else left
        if best > ans:
            ans = best

    return ans

```
