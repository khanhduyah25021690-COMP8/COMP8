```python
#  Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int
def Max_Of_Two(a: int, b: int) -> int:
    if a > b:
        return a
    else:
        return b
print(Max_Of_Two(1, 2))
```

```python
# Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:
from typing import Tuple
def swap(a: int, b: int) -> Tuple[int, int]:
    return b, a
swap (2,1)
```

```python
# Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
is_prime (17)
```

```python
# VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên
def is_perfect_number(n: int) -> bool:
    if n < 2:
        return False
    divisors_sum = 1
    for i in range(2, n // 2 + 1):
        if n % i == 0:
            divisors_sum += i
    return divisors_sum == n
is_perfect_number (23)
```

```python
# #Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k
# trong list số nguyên, nếu không có thì trả về -1
def find_first_index(lst: list[int], k: int) -> int:
    for i, value in enumerate(lst):
        if value == k:
            return i
    return -1
find_first_index([2,45,67,12,90,23], 23)
```

```python
# Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!)
def factorial(n: int) -> int:
    if n == 0:
        return 1
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
factorial(3)

```

```python
#  Viết hàm giả lập một máy tính bỏ túi.
#  Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat \)
#  Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
#  Phép trừ \( - \)
#  Phép cộng \( + \)
#  Phép chia \( / \)
#  Phép nhân \( * \)
#  Kết quả làm tròn đến chữ số thập phân thứ hai.
#  Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
#  Đầu ra: Kết quả phép toán.
def calculator(num1: float, num2: float, operat: str) -> float:
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        result = num1 / num2
    else:
        raise ValueError("Invalid operator")
    return round(result, 2)
```

```python
#  Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví
# dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
#  Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
#  Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu
# có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng
# trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức
# để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal
# distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin
# học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có
# xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương
# pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng
# và thích hợp hơn.
def hamming_distance(x: int, y: int) -> int:
    xor_result = x ^ y
    return bin(xor_result).count("1")
hamming_distance(1,4)

```

```python
#Viết hàm trả về tổng các chữ số của số nguyên dương n
def sum_of_digits(n: int) -> int:
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
sum_of_digits(1234)

```

```python
# Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa
# là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác
# nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là
# hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'.
#  Viết hàm  trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng
# độ dài.
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False

    mapping = {}
    mapped = set()

    for char_a, char_b in zip(a, b):
        if char_a in mapping:
            if mapping[char_a] != char_b:
                return False
        else:
            if char_b in mapped:
                return False
            mapping[char_a] = char_b
            mapped.add(char_b)

    return True
is_isomorphic('egg','add')
```
