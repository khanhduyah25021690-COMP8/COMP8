```python
# W8A1:
import sys

def main():
    for line in sys.stdin:
        line = line.strip()
        if not line:
            continue
        a, b = map(int, line.split())

        try:
            print(f"{a / b:.2f}")
        except ZeroDivisionError:
            print("Loi chia khong")

if __name__ == "__main__":
    main()
```

```python
# W8A2:
import sys

def main():
    # Nhập mảng số nguyên
    my_list = list(map(int, sys.stdin.read().split()))
    n = len(my_list)

    try:
        # Cố tình truy cập chỉ số sai
        print(my_list[n])
    except IndexError:
        print("Truy cap chi so sai")

if __name__ == "__main__":
    main()
```

```python
# W8A3:
import sys

def is_happy(n: int) -> bool:
    seen = set()
    while n != 1:
        if n in seen:
            return False
        seen.add(n)
        n = sum(int(d) ** 2 for d in str(n))
    return True

def main():
    for line in sys.stdin:
        line = line.strip()
        if not line:
            continue

        try:
            a = int(line)
            if a <= 0:
                raise ValueError
        except ValueError:
            print("NO")
            continue

        print("YES" if is_happy(a) else "NO")

if __name__ == "__main__":
    main()
```

```python
# W8A4:
def main():
    try:
        a = int(input())
        b = int(input())
        s = a + b
    except ValueError:
        print("Nhap sai kieu du lieu")
    else:
        print(s)
    finally:
        print("Ket thuc chuong trinh.")

if __name__ == "__main__":
    main()
```

```python
# W8A5:
import sys

def main():
    data = sys.stdin.read().split()
    result = []

    for x in data:
        try:
            result.append(int(x))
        except ValueError:
            print(f"Khong the chuyen '{x}' sang int, bo qua")

    print(result)

if __name__ == "__main__":
    main()
```

```python
# W10A6:
import sys
import math

def main():
    for line in sys.stdin:
        line = line.strip()
        if not line:
            continue
        try:
            x = float(line)
            if x < 0:
                raise ValueError
            print(f"{math.sqrt(x):.2f}")
        except ValueError:
            print("So am")

if __name__ == "__main__":
    main()
```

```python
# W10A8:
import sys

def main():
    for line in sys.stdin:
        filename = line.strip()
        if not filename:
            continue

        try:
            if '.' not in filename:
                raise ValueError

            ext = filename.rsplit('.', 1)[1].lower()
            if ext in ('txt', 'zip'):
                print("Doc file thanh cong")
            else:
                raise ValueError
        except ValueError:
            print("File khong hop le")

if __name__ == "__main__":
    main()
```

```python
# W10A10:
import sys

def main():
    lines = [line.strip() for line in sys.stdin if line.strip() != ""]
    i = 0

    while i < len(lines):
        try:
            data = list(map(int, lines[i].split()))
            i += 1
        except ValueError:
            print("Mang khong hop le")
            continue

        if not data:
            print("Mang khong hop le")
            continue

        N = data[0]
        arr = data[1:]

        if N <= 0 or len(arr) != N:
            print("Mang khong hop le")
            continue

        if len(set(arr)) != N:
            print("Mang khong hop le")
        else:
            print(sum(arr))

if __name__ == "__main__":
    main()
```

```python
# W8A11:
import re
import sys

def main():
    lines = sys.stdin.read().splitlines()
    if not lines:
        return

    t = int(lines[0].strip())
    out = []

    for i in range(1, t + 1):
        s = lines[i].rstrip("\n")  # giữ nguyên backslash nếu có
        try:
            re.compile(s)
            out.append("True")
        except re.error:
            out.append("False")

    print("\n".join(out))

if __name__ == "__main__":
    main()

```
