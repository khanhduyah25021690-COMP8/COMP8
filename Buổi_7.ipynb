```python
# W7A1:
def binary_search(arr: list[int], left: int, right: int, target: int) -> int:
    while left <= right:
        mid = (left + right) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1


if __name__ == "__main__":
    arr = list(map(int, input().split()))
    target = int(input())

    print(binary_search(arr, 0, len(arr) - 1, target))

```

```python
# W7A2
def count_occurrences(arr: list[int], x: int) -> int:
    count = 0
    for v in arr:
        if v == x:
            count += 1
    return count


if __name__ == "__main__":
    arr = list(map(int, input().split()))
    x = int(input())

    print(count_occurrences(arr, x))
```

```python
# W7A3:
def bubble_sort(arr: list[int]) -> list[int]:
    n = len(arr)
    a = arr.copy()  # không làm thay đổi list gốc

    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]

    return a


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(bubble_sort(nums))
```

```python
# W7A4:
def most_frequent_first(arr: list[int]) -> tuple[int, int]:
    freq: dict[int, int] = {}
    first_index: dict[int, int] = {}

    for i, x in enumerate(arr):
        if x not in freq:
            freq[x] = 0
            first_index[x] = i
        freq[x] += 1

    max_count = max(freq.values())

    # Lấy các số có tần suất lớn nhất
    candidates = [x for x, c in freq.items() if c == max_count]

    # Chọn số có vị trí xuất hiện đầu tiên nhỏ nhất
    best = min(candidates, key=lambda x: first_index[x])

    return best, max_count


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    value, count = most_frequent_first(nums)
    print(f"{value} xuat hien nhieu nhat, som nhat, {count} lan")
```

```python
# W7A5:
def count_pairs_with_sum(arr: list[int], X: int) -> int:
    freq: dict[int, int] = {}
    count = 0

    for x in arr:
        y = X - x

        if y in freq:
            count += freq[y]

        freq[x] = freq.get(x, 0) + 1

    return count


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    X = int(input())

    print(count_pairs_with_sum(nums, X))
```

```python
# W7A6:
def flatten_list(nested: list[list[int]]) -> list[int]:
    result = []

    for sublist in nested:
        for x in sublist:
            result.append(x)

    return result


if __name__ == "__main__":
    nested = eval(input())
    print(flatten_list(nested))
```

```python
# W7A7:
def length_of_lis(nums: list[int]) -> int:
    if not nums:
        return 0

    n = len(nums)
    dp = [1] * n  # dp[i] = độ dài LIS kết thúc tại i

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)


if __name__ == "__main__":
    nums = eval(input())   # input dạng [10, 9, 2, 5, 3, 7, 101, 18]
    print(length_of_lis(nums))
```

```python
# W7A8:
import heapq


def min_interval(intervals: list[list[int]], queries: list[int]) -> list[int]:
    # Sắp xếp intervals theo l tăng dần
    intervals.sort()

    # Gắn chỉ số cho queries để trả kết quả đúng thứ tự
    indexed_queries = sorted((q, i) for i, q in enumerate(queries))

    result = [-1] * len(queries)
    heap = []  # (length, r)
    i = 0

    for q, idx in indexed_queries:
        # Đưa vào heap các interval có l <= q
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1

        # Loại bỏ các interval không còn chứa q
        while heap and heap[0][1] < q:
            heapq.heappop(heap)

        if heap:
            result[idx] = heap[0][0]

    return result


if __name__ == "__main__":
    intervals = eval(input())
    queries = eval(input())

    print(min_interval(intervals, queries))
```

```python
# W7A9:
def count_smaller(nums: list[int]) -> list[int]:
    # Nén giá trị (coordinate compression)
    sorted_unique = sorted(set(nums))
    rank = {v: i + 1 for i, v in enumerate(sorted_unique)}

    size = len(sorted_unique)
    bit = [0] * (size + 1)

    def update(i):
        while i <= size:
            bit[i] += 1
            i += i & -i

    def query(i):
        s = 0
        while i > 0:
            s += bit[i]
            i -= i & -i
        return s

    result = [0] * len(nums)

    # Duyệt từ phải sang trái
    for i in range(len(nums) - 1, -1, -1):
        r = rank[nums[i]]
        result[i] = query(r - 1)  # đếm số < nums[i]
        update(r)

    return result


if __name__ == "__main__":
    nums = eval(input())
    print(count_smaller(nums))
```

```python
# W7A10:
import bisect


def max_subarray_sum_mod(nums: list[int], m: int) -> int:
    prefix = 0
    best = 0
    sorted_prefixes = []

    for x in nums:
        prefix = (prefix + x) % m

        # Tìm prefix nhỏ nhất > prefix hiện tại
        idx = bisect.bisect_right(sorted_prefixes, prefix)

        if idx < len(sorted_prefixes):
            best = max(best, (prefix - sorted_prefixes[idx] + m) % m)

        best = max(best, prefix)

        bisect.insort(sorted_prefixes, prefix)

    return best


if __name__ == "__main__":
    nums = eval(input())   # input dạng [3, 3, 9, 9, 5]
    m = int(input())

    print(max_subarray_sum_mod(nums, m))
```

```python
# W7A11:
import re


def count_keyword(text: str, keyword: str) -> int:
    # \b = word boundary, đảm bảo khớp nguyên từ
    pattern = r"\b" + re.escape(keyword) + r"\b"
    matches = re.findall(pattern, text)
    return len(matches)


if __name__ == "__main__":
    text = input().strip()
    keyword = input().strip()

    print(count_keyword(text, keyword))
```

```python
# W7A12:
def fastest_finger_first(options: list[str]) -> str:
    labels = ["A", "B", "C", "D"]

    # Gắn mỗi cụm từ với nhãn tương ứng
    paired = list(zip(options, labels))

    # Sắp xếp theo cụm từ (từ điển tăng dần)
    paired.sort(key=lambda x: x[0])

    # Lấy ra thứ tự nhãn sau khi sắp xếp
    result = [label for _, label in paired]

    return " ".join(result)


if __name__ == "__main__":
    options = [input().strip() for _ in range(4)]
    print(fastest_finger_first(options))
```

```python
# W7A13:
from collections import Counter


def longest_harmonious_subsequence(arr: list[int]) -> int:
    freq = Counter(arr)
    best = 0

    for x in freq:
        if x + 1 in freq:
            best = max(best, freq[x] + freq[x + 1])

    return best


if __name__ == "__main__":
    arr = list(map(int, input().split()))
    print(longest_harmonious_subsequence(arr))
```

```python
# W7A14:
def find_seven_positions(nums: list[int]) -> list[int]:
    positions = []

    for i, x in enumerate(nums):
        if x == 7:
            positions.append(i)

    return positions


if __name__ == "__main__":
    n = int(input())
    nums = list(map(int, input().split()))

    result = find_seven_positions(nums)

    if result:
        print(" ".join(map(str, result)))
    else:
        print("Not found")
```
