```python
# W8A1:
class Cylinder:
    PI = 3.14

    def __init__(self, radius: float, height: float):
        self.radius = radius
        self.height = height

    def surface_area(self) -> float:
        r = self.radius
        h = self.height
        return 2 * self.PI * r * r + 2 * self.PI * r * h

    def volume(self) -> float:
        r = self.radius
        h = self.height
        return self.PI * r * r * h


if __name__ == "__main__":
    radius, height = map(float, input().split())

    cylinder = Cylinder(radius, height)

    area = cylinder.surface_area()
    vol = cylinder.volume()

    print(f"{area:.2f}")
    print(f"{vol:.2f}")
```

```python
# W8A2:
class Date:
    def __init__(self, day: int, month: int, year: int):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self) -> bool:
        y = self.year
        return (y % 400 == 0) or (y % 4 == 0 and y % 100 != 0)

    def days_in_month(self) -> int:
        if self.month < 1 or self.month > 12:
            return 0

        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2:
            return 29 if self.is_leap_year() else 28

    def is_valid(self) -> bool:
        if self.year < 1:
            return False
        if self.month < 1 or self.month > 12:
            return False
        max_day = self.days_in_month()
        return 1 <= self.day <= max_day

    def next_day(self):
        if not self.is_valid():
            return None

        d, m, y = self.day, self.month, self.year

        if d < self.days_in_month():
            d += 1
        else:
            d = 1
            if m < 12:
                m += 1
            else:
                m = 1
                y += 1

        return Date(d, m, y)

    def __str__(self) -> str:
        return f"{self.day:02d}/{self.month:02d}/{self.year:04d}"


if __name__ == "__main__":
    s = input().strip()

    try:
        day, month, year = map(int, s.split("/"))
    except ValueError:
        print("INVALID")
        exit()

    date = Date(day, month, year)

    if not date.is_valid():
        print("INVALID")
    else:
        next_date = date.next_day()
        print(next_date)
```

```python
# W8A3:
import math


class Point:
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y

    def is_origin(self) -> bool:
        return self.x == 0 and self.y == 0

    def on_x_axis(self) -> bool:
        return self.y == 0 and self.x != 0

    def on_y_axis(self) -> bool:
        return self.x == 0 and self.y != 0

    def distance_to_origin(self) -> float:
        return math.sqrt(self.x ** 2 + self.y ** 2)


if __name__ == "__main__":
    x, y = map(float, input().split())
    point = Point(x, y)

    # Kiểm tra gốc tọa độ
    if point.is_origin():
        print("M is at the origin")
    else:
        print("M is not at the origin")

    # Kiểm tra trục tung
    if point.on_y_axis():
        print("M is on the y-axis")
    else:
        print("M is not on the y-axis")

    # Kiểm tra trục hoành
    if point.on_x_axis():
        print("M is on the x-axis")
    else:
        print("M is not on the x-axis")

    # Khoảng cách tới gốc tọa độ
    distance = point.distance_to_origin()
    print(f"{distance:.2f}")
```

```python
# W8A4:
class Calculator:
    def __init__(self, a: float, b: float):
        self.a = a
        self.b = b

    def set_numbers(self, a: float, b: float):
        self.a = a
        self.b = b

    def add(self):
        return self.a + self.b

    def subtract(self):
        return self.a - self.b

    def product(self):
        return self.a * self.b

    def divide(self):
        if self.b == 0:
            return "Error: Division by zero"
        return self.a / self.b

    def power(self):
        return self.a ** self.b

    def mod(self):
        if self.b == 0:
            return "Error: Modulo by zero"
        return self.a % self.b


def main():
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))

    calc = Calculator(a, b)

    while True:
        print("\n--- Calculator Menu ---")
        print("1. Add")
        print("2. Subtract")
        print("3. Product")
        print("4. Divide")
        print("5. Power")
        print("6. Mod")
        print("7. Set new numbers")

        choice = input("Choose an option (1-7): ").strip()

        if choice == "1":
            print("Result:", calc.add())

        elif choice == "2":
            print("Result:", calc.subtract())

        elif choice == "3":
            print("Result:", calc.product())

        elif choice == "4":
            print("Result:", calc.divide())

        elif choice == "5":
            print("Result:", calc.power())

        elif choice == "6":
            print("Result:", calc.mod())

        elif choice == "7":
            a = float(input("Enter first number: "))
            b = float(input("Enter second number: "))
            calc.set_numbers(a, b)
            print("Numbers updated!")

        else:
            print("Invalid choice!")

        exit_choice = input("Do you want to exit? (yes/YES/Yes): ").strip()
        if exit_choice.lower() == "yes":
            print("Goodbye!")
            break


if __name__ == "__main__":
    main()
```

```python
W8A5:
class ShoppingCart:
    def __init__(self):
        self.items = []   # mỗi phần tử là dict {"name": ..., "price": ...}

    def add_product(self, name: str, price: float):
        self.items.append({"name": name, "price": price})
        print("Product added!")

    def remove_product(self, name: str):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                print("Product removed!")
                return
        print("Product not found!")

    def is_empty(self) -> bool:
        return len(self.items) == 0

    def total_price(self) -> float:
        return sum(item["price"] for item in self.items)

    def show_cart(self):
        if self.is_empty():
            print("Cart is empty!")
        else:
            print("\n--- Shopping Cart ---")
            for i, item in enumerate(self.items, start=1):
                print(f"{i}. {item['name']} - {item['price']}")
            print("----------------------")

    def clear_cart(self):
        self.items.clear()
        print("Cart cleared!")


def main():
    cart = ShoppingCart()

    while True:
        print("\n--- Shopping Cart Menu ---")
        print("2. Add product")
        print("3. Remove product")
        print("4. Check if cart is empty")
        print("5. Total price")
        print("6. Show cart")
        print("7. Clear cart")

        choice = input("Choose an option (2-7): ").strip()

        if choice == "2":
            name = input("Enter product name: ").strip()
            price = float(input("Enter product price: "))
            cart.add_product(name, price)

        elif choice == "3":
            name = input("Enter product name to remove: ").strip()
            cart.remove_product(name)

        elif choice == "4":
            if cart.is_empty():
                print("Cart is empty!")
            else:
                print("Cart is not empty!")

        elif choice == "5":
            print("Total price:", cart.total_price())

        elif choice == "6":
            cart.show_cart()

        elif choice == "7":
            cart.clear_cart()

        else:
            print("Invalid choice!")

        exit_choice = input("Do you want to exit? (yes): ").strip()
        if exit_choice.lower() == "yes":
            print("Goodbye!")
            break


if __name__ == "__main__":
    main()
```

```python
# W8A6:
class Rectangle:
    def __init__(self, w: float, h: float):
        self.w = w
        self.h = h

    def area(self) -> float:
        return self.w * self.h

    def perimeter(self) -> float:
        return 2 * (self.w + self.h)

    def scale(self, k: float):
        self.w *= k
        self.h *= k


if __name__ == "__main__":
    w, h, k = map(float, input().split())

    rect = Rectangle(w, h)
    rect.scale(k)

    area = rect.area()
    peri = rect.perimeter()

    print(f"{area:.2f} {peri:.2f}")
```

```python
# W8A7:
import math


class Fraction:
    def __init__(self, num: int, den: int):
        if den == 0:
            raise ValueError("Denominator cannot be zero")

        # Quy ước mẫu dương
        if den < 0:
            num = -num
            den = -den

        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g

    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)

    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)

    def div(self, other):
        if other.num == 0:
            raise ValueError("Division by zero fraction")

        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)

    def __str__(self):
        return f"{self.num}/{self.den}"


if __name__ == "__main__":
    a, b, op, c, d = input().split()
    a, b, c, d = int(a), int(b), int(c), int(d)

    f1 = Fraction(a, b)
    f2 = Fraction(c, d)

    if op == "+":
        result = f1.add(f2)
    elif op == "-":
        result = f1.sub(f2)
    elif op == "*":
        result = f1.mul(f2)
    elif op == "/":
        result = f1.div(f2)
    else:
        print("Invalid operator")
        exit()

    print(result)
```

```python
# W8A8:
class BankAccount:
    def __init__(self, owner: str, balance: float):
        self.owner = owner
        self.balance = balance

    def deposit(self, x: float):
        self.balance += x

    def withdraw(self, x: float):
        if x <= self.balance:
            self.balance -= x
        # nếu rút quá số dư thì bỏ qua giao dịch

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"


if __name__ == "__main__":
    owner, initial_balance = input().split()
    balance = float(initial_balance)

    account = BankAccount(owner, balance)

    n = int(input())

    for _ in range(n):
        parts = input().split()
        cmd = parts[0]
        amount = float(parts[1])

        if cmd == "DEPOSIT":
            account.deposit(amount)
        elif cmd == "WITHDRAW":
            account.withdraw(amount)

    # In số dư cuối cùng
    print(f"{account.balance:.2f}")
```

```python
# W8A10:
class Book:
    def __init__(self, title: str, author: str, year: int):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book: Book):
        self.books.append(book)

    def count_by_author(self, name: str) -> int:
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y: int) -> int:
        return sum(1 for b in self.books if b.year == y)


if __name__ == "__main__":
    n = int(input())
    library = Library()

    for _ in range(n):
        line = input().strip()

        if line.startswith("ADD"):
            _, rest = line.split(" ", 1)
            title, author, year = rest.split(";")
            book = Book(title, author, int(year))
            library.add(book)

        elif line.startswith("COUNTYEAR"):
            _, year = line.split()
            print(library.find_by_year(int(year)))

        elif line.startswith("COUNT"):
            _, author = line.split()
            print(library.count_by_author(author))
```

```python
# W8A11:
import math


def vector_sub(a, b):
    return (a[0] - b[0], a[1] - b[1], a[2] - b[2])


def cross(u, v):
    return (
        u[1] * v[2] - u[2] * v[1],
        u[2] * v[0] - u[0] * v[2],
        u[0] * v[1] - u[1] * v[0],
    )


def dot(u, v):
    return u[0] * v[0] + u[1] * v[1] + u[2] * v[2]


def norm(u):
    return math.sqrt(dot(u, u))


if __name__ == "__main__":
    A = tuple(map(float, input().split()))
    B = tuple(map(float, input().split()))
    C = tuple(map(float, input().split()))
    D = tuple(map(float, input().split()))

    AB = vector_sub(B, A)
    AC = vector_sub(C, A)
    BC = vector_sub(C, B)
    BD = vector_sub(D, B)

    n1 = cross(AB, AC)
    n2 = cross(BC, BD)

    cos_phi = abs(dot(n1, n2)) / (norm(n1) * norm(n2))
    cos_phi = max(-1.0, min(1.0, cos_phi))  # tránh lỗi số học

    phi = math.degrees(math.acos(cos_phi))

    print(f"{phi:.2f}")
```

```python
# W8A12:
class ComplexNumber:
    def __init__(self, real_part: float, imaginary_part: float):
        self.real = real_part
        self.imag = imaginary_part

    def multiply(self, other):
        real = self.real * other.real - self.imag * other.imag
        imag = self.real * other.imag + self.imag * other.real
        return ComplexNumber(real, imag)


if __name__ == "__main__":
    a, b = map(float, input().split())
    c, d = map(float, input().split())

    z1 = ComplexNumber(a, b)
    z2 = ComplexNumber(c, d)

    result = z1.multiply(z2)

    # In phần thực và phần ảo
    if result.real.is_integer():
        real_out = int(result.real)
    else:
        real_out = result.real

    if result.imag.is_integer():
        imag_out = int(result.imag)
    else:
        imag_out = result.imag

    print(real_out, imag_out)
```

```python
# W8A13:
import math


class Triangle:
    def __init__(self, edge1: int, edge2: int, edge3: int):
        self.a = edge1
        self.b = edge2
        self.c = edge3

    def is_valid(self) -> bool:
        return (
            self.a + self.b > self.c and
            self.a + self.c > self.b and
            self.b + self.c > self.a
        )

    def getArea(self) -> float:
        if not self.is_valid():
            return -1

        p = (self.a + self.b + self.c) / 2
        area = math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
        return area


if __name__ == "__main__":
    a, b, c = map(int, input().split())

    tri = Triangle(a, b, c)
    area = tri.getArea()

    if area == -1:
        print("invalid")
    else:
        # In ra diện tích, làm tròn 2 chữ số
        if area.is_integer():
            print(int(area))
        else:
            print(f"{area:.2f}")
```

```python
# W8A9:
class Student:
    def __init__(self, name: str):
        self.name = name
        self.scores = []

    def add_score(self, subject: str, score: float):
        self.scores.append(score)

    def avg(self) -> float:
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self) -> str:
        a = self.avg()
        if a >= 8:
            return "Excellent"
        elif a >= 6.5:
            return "Good"
        elif a >= 5:
            return "Average"
        else:
            return "Poor"


if __name__ == "__main__":
    name = input().strip()
    student = Student(name)

    n = int(input())

    for _ in range(n):
        subject, score = input().split()
        student.add_score(subject, float(score))

    average = student.avg()
    ranking = student.rank()

    print(f"{student.name} {average:.2f} {ranking}")
```
