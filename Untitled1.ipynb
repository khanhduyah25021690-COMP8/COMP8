```python
# W2A3:
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))
print("Tổng:", a + b)
print("Hiệu:", a - b)
print("Tích:", a * b)
print("Phần nguyên:", a // b)
print("Phần dư:", a % b)
print("Chia thực:", round(a / b, 2))
```
