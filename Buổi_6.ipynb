```python
# W6A1
def remove_duplicates_keep_order(nums: list[int]) -> list[int]:
    seen = set()
    result = []

    for x in nums:
        if x not in seen:
            seen.add(x)
            result.append(x)

    return result


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(remove_duplicates_keep_order(nums))
```

```python
# W6A2
def prefix_sums(nums: list[int]) -> list[int]:
    running_sum = 0
    result = []

    for x in nums:
        running_sum += x
        result.append(running_sum)

    return result


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(prefix_sums(nums))
```

```python
# W6A3:
def rotate_left_as_tuple(nums: list[int], k: int) -> tuple[int, ...]:
    n = len(nums)
    if n == 0:
        return ()

    k %= n  # đảm bảo k nằm trong [0, n-1]
    rotated = nums[k:] + nums[:k]
    return tuple(rotated)


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    k = int(input())
    print(rotate_left_as_tuple(nums, k))
```

```python
# W6A4:
def group_values_by_key(pairs_line: str) -> dict[str, list[str]]:
    result: dict[str, list[str]] = {}

    for token in pairs_line.split():
        key, value = token.split(":", 1)  # tách đúng 1 dấu ':' đầu tiên

        if key not in result:
            result[key] = []

        result[key].append(value)

    return result


if __name__ == "__main__":
    line = input().strip()
    print(group_values_by_key(line))

```

```python
# W6A5
def count_signs(nums: list[int]) -> dict[str, int]:
    counts = {"positives": 0, "negatives": 0, "zeros": 0}

    for x in nums:
        if x > 0:
            counts["positives"] += 1
        elif x < 0:
            counts["negatives"] += 1
        else:
            counts["zeros"] += 1

    return counts


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(count_signs(nums))
```

```python
# W6A7
def process_tuple(items: list[str]) -> tuple[str, str, tuple[str, ...]]:
    first = items[0]
    last = items[-1]
    reversed_tuple = tuple(reversed(items))

    return first, last, reversed_tuple


if __name__ == "__main__":
    items = input().split()
    first, last, reversed_tuple = process_tuple(items)

    print(first, last, reversed_tuple)
```

```python
# W6A8
def count_strings(words: list[str]) -> dict[str, int]:
    counts: dict[str, int] = {}

    for w in words:
        if w not in counts:
            counts[w] = 0
        counts[w] += 1

    return counts


if __name__ == "__main__":
    words = input().split()
    print(count_strings(words))
```

```python
# W6A9
def merge_and_sum_dicts(d1: dict[str, int], d2: dict[str, int]) -> dict[str, int]:
    merged: dict[str, int] = {}
    all_keys = set(d1.keys()) | set(d2.keys())
    for key in sorted(all_keys):
        value1 = d1.get(key, 0)
        value2 = d2.get(key, 0)
        merged[key] = value1 + value2

    return merged


if __name__ == "__main__":
    d1 = eval(input())
    d2 = eval(input())

    print(merge_and_sum_dicts(d1, d2))
```

```python
# W6A10
def find_pairs_with_sum(nums: list[int], k: int) -> list[tuple[int, int]]:
    seen = {}
    pairs = []

    for x in nums:
        y = k - x

        if y in seen:
            pairs.append((y, x))

        seen[x] = seen.get(x, 0) + 1

    pairs.sort(key=lambda t: t[0])
    return pairs


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    k = int(input())

    print(find_pairs_with_sum(nums, k))
```

```python
# W6A11
def split_even_odd(nums: list[int]) -> tuple[tuple[int, ...], tuple[int, ...]]:
    evens = []
    odds = []

    for x in nums:
        if x % 2 == 0:
            evens.append(x)
        else:
            odds.append(x)

    return tuple(evens), tuple(odds)


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    evens, odds = split_even_odd(nums)

    print(evens)
    print(odds)
```

```python
# W6A12
def most_frequent_smallest(nums: list[int]) -> int:
    freq: dict[int, int] = {}

    # Đếm tần suất
    for x in nums:
        freq[x] = freq.get(x, 0) + 1

    # Tìm tần suất lớn nhất
    max_count = max(freq.values())

    # Lấy các số có cùng tần suất lớn nhất
    candidates = [x for x, count in freq.items() if count == max_count]

    # Trả về số nhỏ nhất trong các ứng viên
    return min(candidates)


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(most_frequent_smallest(nums))
```

```python
# W6A13
def invert_dict_from_line(line: str) -> dict[str, str]:
    tokens = line.split()
    inverted: dict[str, str] = {}

    for i in range(0, len(tokens), 2):
        key = tokens[i]
        value = tokens[i + 1]
        inverted[value] = key

    return inverted


if __name__ == "__main__":
    line = input().strip()
    print(invert_dict_from_line(line))
```

```python
# W6A14
def intersection_lists(nums1: list[int], nums2: list[int]) -> list[int]:
    set2 = set(nums2)
    result = []
    seen = set()

    for x in nums1:
        if x in set2 and x not in seen:
            result.append(x)
            seen.add(x)

    return result


if __name__ == "__main__":
    nums1 = list(map(int, input().split()))
    nums2 = list(map(int, input().split()))

    print(intersection_lists(nums1, nums2))
```

```python
# W6A15
def filter_dict_by_value(line: str, k: int) -> dict[str, int]:
    tokens = line.split()
    result: dict[str, int] = {}

    for i in range(0, len(tokens), 2):
        key = tokens[i]
        value = int(tokens[i + 1])

        if value > k:
            result[key] = value

    return result


if __name__ == "__main__":
    line = input().strip()
    k = int(input())

    print(filter_dict_by_value(line, k))
```

```python
# W6A17
def print_diagonals(matrix: list[list[int]]) -> None:
    n = len(matrix)

    # Đường chéo chính
    for i in range(n):
        print(matrix[i][i], end=" ")
    print()

    # Đường chéo phụ
    for i in range(n):
        print(matrix[i][n - 1 - i], end=" ")
    print()


if __name__ == "__main__":
    n = int(input())

    matrix = []
    for _ in range(n):
        row = list(map(int, input().split()))
        matrix.append(row)

    print_diagonals(matrix)
```

```python
# W6A19
def index_of_max(nums: list[int]) -> int:
    max_value = nums[0]
    max_index = 0

    for i in range(1, len(nums)):
        if nums[i] > max_value:
            max_value = nums[i]
            max_index = i

    return max_index + 1  # đổi sang chỉ số bắt đầu từ 1


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    print(index_of_max(nums))

```

```python
# W6A20
def first_index_of_k(nums: list[int], k: int) -> int:
    for i in range(len(nums)):
        if nums[i] == k:
            return i + 1  # chỉ số bắt đầu từ 1
    return -1


if __name__ == "__main__":
    nums = list(map(int, input().split()))
    k = int(input())

    print(first_index_of_k(nums, k))
```

```python
# W6A21
def map_strings_to_positions(words: list[str]) -> dict[str, tuple[int, ...]]:
    positions: dict[str, list[int]] = {}

    for i, w in enumerate(words):
        if w not in positions:
            positions[w] = []
        positions[w].append(i)

    # Chuyển list vị trí sang tuple
    result = {k: tuple(v) for k, v in positions.items()}
    return result


if __name__ == "__main__":
    words = input().split()
    print(map_strings_to_positions(words))
```
