```python
#  W4A1:Viết chương trình Python sử dụng một trong 2 loại vòng lặp
# for hoặc while để tính tống các số nguyên từ 1 đến n. (n <=
# 1000)
# Solution 1:
n = int(input("Nhập số nguyên n (≤ 1000): "))
if n <= 1000:
    tong = 0
    for i in range(1, n + 1):
        tong += i
    print("Tổng các số từ 1 đến", n, "là:", tong)
else:
    print("Giá trị n vượt quá giới hạn cho phép.")
# Solution 2:
n = int(input("Nhập số nguyên n (≤ 1000): "))
if n <= 1000:
    tong = 0
    i = 1
    while i <= n:
        tong += i
        i += 1
    print("Tổng các số từ 1 đến", n, "là:", tong)
else:
    print("Giá trị n vượt quá giới hạn cho phép.")

```

```python
# W4A6: Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số
# nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)."
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

while True:
    try:
        line = input("Nhập a b (hoặc Enter để kết thúc): ").strip()
        if line == "":
            break
        a, b = map(int, line.split())
        total = 0
        for num in range(a, b + 1):
            if is_prime(num):
                total += num
        print(f"Tổng các số nguyên tố trong đoạn [{a}, {b}] là: {total}")
    except ValueError:
        print("Dữ liệu không hợp lệ, vui lòng nhập lại.")
```

```python
# W4A7: Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số
# nguyên tố lớn nhất của n."
def largest_prime_factor(n):
    factor = 2
    largest = 1
    while n > 1:
        while n % factor == 0:
            largest = factor
            n //= factor
        factor += 1
    return largest

n = int(input("Nhập số nguyên dương n (n ≥ 2): "))
print(f"Ước số nguyên tố lớn nhất của {n} là: {largest_prime_factor(n)}")
```

```python
# W4A8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse
#  (n) cho đến khi n là số đối xứng (palindrome). In ra số bước
# và giá trị palindrome thu được.
def reverse_number(x):
    return int(str(x)[::-1])

def is_palindrome(x):
    return str(x) == str(x)[::-1]

n = int(input("Nhập số nguyên dương n: "))

steps = 0
while not is_palindrome(n):
    n = n + reverse_number(n)
    steps += 1

print(f"Số bước: {steps}")
print(f"Palindrome thu được: {n}")
```

```python
# W4A9: Nhập vào số nguyên dương n. In ra tất cả các số chính
# phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự
# tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu
# không có, in ra (no number).
import math

def distinct_digits(num):
    digits = str(num)
    return len(set(digits)) == len(digits)

n = int(input("Nhập số nguyên dương n: "))

results = []
for i in range(1, int(math.isqrt(n)) + 1):
    sq = i * i
    if sq <= n and distinct_digits(sq):
        results.append(sq)

if results:
    print(" ".join(map(str, results)))
else:
    print("(no number)")
```

```python
# W4A10:Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài
# dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x
# L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất."
def collatz_length(x, memo):
    if x == 1:
        return 1
    if x in memo:
        return memo[x]
    if x % 2 == 0:
        length = 1 + collatz_length(x // 2, memo)
    else:
        length = 1 + collatz_length(3 * x + 1, memo)
    memo[x] = length
    return length

n = int(input("Nhập số nguyên dương n: "))

memo = {}
best_x, best_len = 1, 1
for i in range(1, n + 1):
    L = collatz_length(i, memo)
    if L > best_len or (L == best_len and i < best_x):
        best_x, best_len = i, L

print(f"({best_x}, {best_len})")
```

```python
# W4A11: Viết chương trình đếm số lượng ước số chẵn của số nguyên
# dương n với n < 10^6.
import math

n = int(input("Nhập số nguyên dương n (< 10^6): "))
count = 0

for i in range(1, int(math.isqrt(n)) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1
        if (n // i != i) and (n // i % 2 == 0):
            count += 1

print(f"Số lượng ước số chẵn của {n} là: {count}")

```

```python
# W4A13: Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là
# cặp số thân thiết không. Nếu có, in ra màn hình "true",
# ngược lại, in ra "false"
def sum_proper_divisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

if sum_proper_divisors(a) == b and sum_proper_divisors(b) == a:
    print("true")
else:
    print("false")
```

```python
# W4A14: Cho hai số m và n là hai số nguyên dương nhập từ bàn
# phím. Viết chương trình tìm ước chung lớn nhất của hai số
# m và n
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

m = int(input("Nhập số nguyên dương m: "))
n = int(input("Nhập số nguyên dương n: "))

print(f"Ước chung lớn nhất của {m} và {n} là: {gcd(m, n)}")
```

```python
# W4A15:  [ChickenAndDog]
#  "Vừa gà vừa chó,
#  Bó lại cho tròn,
#  Ba mươi sáu con,
#  Một trăm chân chẵn"
#  Từ bài toán dân gian trên, mở rộng thành chương trình nhận
# đầu vào là tổng số con và tổng số chân của gà và chó. Nếu
# tìm được số phù hợp, in ra số lượng gà và số lượng chó.
# Ngược lại, in ra "invalid"
total_animals = int(input("Nhập tổng số con: "))
total_legs = int(input("Nhập tổng số chân: "))

found = False
for chickens in range(total_animals + 1):
    dogs = total_animals - chickens
    if 2 * chickens + 4 * dogs == total_legs:
        print(f"Số gà: {chickens}, Số chó: {dogs}")
        found = True
        break

if not found:
    print("invalid")
```

```python
# W4A20:  Nhập vào một số nguyên dương n, kiểm tra xem n có phải là
# số dạng 2^k hay không
n = int(input("Nhập số nguyên dương n: "))

def is_power_of_two(n):
    return n > 0 and (n & (n - 1)) == 0

if is_power_of_two(n):
    print("n là số dạng 2^k")
else:
    print("n không phải là số dạng 2^k")
```

```python
# W4A22:  Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ
# số chẵn, bao nhiêu chữ số lẻ
n = int(input("Nhập số nguyên dương n: "))
even_count = 0
odd_count = 0

while n > 0:
    digit = n % 10
    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    n //= 10

print("Số chữ số chẵn:", even_count)
print("Số chữ số lẻ:", odd_count)
```

```python
# W4A23:  Nhập n
#  Cho S(k) = 1 + 2 + 3 + … + k
#  Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n
n = int(input("Nhập số nguyên dương n: "))

k = 1
while (k * (k + 1)) // 2 < n:
    k += 1

print("Giá trị k lớn nhất là:", k - 1)
```

```python
# W4A24:  Nhập vào A, tìm n nhỏ nhất sao cho
#  1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A
A = float(input("Nhập giá trị A: "))
sum = 0.0
n = 1

while sum <= A:
    sum += 1 / n
    n += 1

print("Giá trị n nhỏ nhất là:", n - 1)
```
